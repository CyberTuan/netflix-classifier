```python
df = df[['type', 'country', 'release_year', 'duration']]
df.dropna(inplace=True)
```

```python
df = df[['type', 'country', 'release_year','duration']].copy()
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert categorical to numeric
le_type = LabelEncoder()
le_Country = LabelEncoder()
le_duration = LabelEncoder()


df['type'] = le_type.fit_transform(df['type'])
df['country'] = le_Country.fit_transform(df['country'])
df['duration'] = le_duration.fit_transform(df['duration'])

from sklearn.model_selection import train_test_split
x = df[['country', 'release_year','duration']]
y = df['type']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state= 42)

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

#Create and train model
model = RandomForestClassifier()
model.fit(x_train, y_train)

#Make predicition
y_pred = model.predict(x_test)

#Eval performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=[str(cls) for cls in le_type.classes_]))



```

```python
import joblib
joblib.dump(model, 'netflix_model.pk1')
```
